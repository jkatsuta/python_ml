```python
class LinearRegressionGD(object):
    def __init__(self, eta=0.001, n_iter=20):
        self.eta = eta
        self.n_iter = n_iter
        
    def fit(self, X, y):
        self.w_ = np.zeros(1 + X.shape[1])
        self.cost_ = []
        for i in range(self.n_iter):
            output = self.net_input(X)
            errors = (y.reshape(y.shape[0]) - output)
            self.w_[1:] += self.eta * X.T.dot(errors)
            self.w_[0] += self.eta * errors.sum()
            cost = (errors**2).sum() / 2.
            self.cost_.append(cost)
        return self
    
    def net_input(self, X):
        return np.dot(X, self.w_[1:]) + self.w_[0]
    
    def predict(self, X):
        return self.net_input(X)
        
```

```python
from sklearn.preprocessing import StandardScaler
X = df[['RM']].values
y = df[["MEDV"]].values

sc_x = StandardScaler()
sc_y = StandardScaler()
X_std = sc_x.fit_transform(X)
y_std = sc_y.fit_transform(y)

lr = LinearRegressionGD()
lr.fit(X_std, y_std)
```

```python
plt.plot(range(1, lr.n_iter+1), lr.cost_)
```

```python
def lin_regplot(X, y, model):
    plt.scatter(X, y, c='blue')
    plt.plot(X, model.predict(X), color='red')
```

```python
lin_regplot(X_std, y_std, lr)
print(lr.w_)
```

```python
num_room_std = sc_x.transform([[5.0]])
price_std = lr.predict(num_room_std)
print(price_std, sc_y.inverse_transform(price_std))
```

## 10.4 Robust regression

```python
from sklearn.linear_model import RANSACRegressor
ransac = RANSACRegressor(LinearRegression(),
                         max_trials=100,
                         min_samples=50,
                         loss='absolute_loss',
                         residual_threshold=5.0,
                         random_state=0)
ransac.fit(X, y)
```

```python
inlier_mask = ransac.inlier_mask_
line_X = np.arange(3, 10 ,1)
line_y_ransac = ransac.predict(line_X.reshape(-1, 1))

plt.scatter(X[inlier_mask], y[inlier_mask], c='b', label='inliers')
plt.scatter(X[~inlier_mask], y[~inlier_mask], c='lightgreen', label='outliers')
plt.plot(line_X, line_y_ransac, 'r-')
plt.plot(line_X, slr.predict(line_X.reshape(-1, 1)), 'y--')
plt.legend(loc='best')
```

## 10.5 evaluation

```python
from sklearn.cross_validation import train_test_split
X = df.iloc[:, :-1].values
y = df['MEDV'].values
X_train, X_test, y_train, y_test =\
    train_test_split(X, y, test_size=0.3, random_state=0)
slr = LinearRegression()    
slr.fit(X_train, y_train)
y_train_pred = slr.predict(X_train)
y_test_pred = slr.predict(X_test)
```

```python
plt.scatter(y_train_pred, y_train_pred - y_train, c='b', marker='o')
plt.scatter(y_test_pred, y_test_pred - y_test, c='lightblue', marker='s')
plt.axhline(0, xmin=-10, xmax=50, color='red')
```

```python
from sklearn.metrics import mean_squared_error as mse
print('MSE train:', mse(y_train, y_train_pred), ' MSE test:', mse(y_test, y_test_pred))

from sklearn.metrics import r2_score
print('R^2 train:', r2_score(y_train, y_train_pred), ' R^2 test:', r2_score(y_test, y_test_pred))

# -> suggesting overfit
```

## 10.6 regularization

```python
from sklearn.linear_model import Ridge, Lasso, ElasticNet
ridge = Ridge(alpha=1.0)
#ridge = Lasso(alpha=0.01)
ridge.fit(X_train, y_train)

y_train_pred = ridge.predict(X_train)
y_test_pred = ridge.predict(X_test)

plt.scatter(y_train_pred, y_train_pred - y_train, c='b', marker='o')
plt.scatter(y_test_pred, y_test_pred - y_test, c='lightblue', marker='s')
plt.axhline(0, xmin=-10, xmax=50, color='red')

from sklearn.metrics import mean_squared_error as mse
print('MSE train:', mse(y_train, y_train_pred), ' MSE test:', mse(y_test, y_test_pred))

from sklearn.metrics import r2_score
print('R^2 train:', r2_score(y_train, y_train_pred), ' R^2 test:', r2_score(y_test, y_test_pred))
```

## 10.7 Polynomial

```python
from sklearn.preprocessing import PolynomialFeatures
X = df[['LSTAT']].values
y = df['MEDV'].values

regr = LinearRegression()

quadratic = PolynomialFeatures(degree=2)
cubic = PolynomialFeatures(degree=3)
X_quad = quadratic.fit_transform(X)
X_cubic = cubic.fit_transform(X)
X_fit = np.arange(X.min(), X.max(), 1).reshape(-1, 1)

regr.fit(X, y)
y_lin_fit = regr.predict(X_fit)
lin_r2 = r2_score(y, regr.predict(X))

regr.fit(X_quad, y)
y_quad_fit = regr.predict(quadratic.fit_transform(X_fit))
quad_r2 = r2_score(y, regr.predict(X_quad))

regr.fit(X_cubic, y)
y_cubic_fit = regr.predict(cubic.fit_transform(X_fit))
cubic_r2 = r2_score(y, regr.predict(X_cubic))

plt.scatter(X, y, c='lightgray', label='train data')
plt.plot(X_fit, y_lin_fit, 'k-', label='linear(d=1), $R^2=%.2f$' % lin_r2)
plt.plot(X_fit, y_quad_fit, 'g-', label='quadratic(d=2), $R^2=%.2f$' % quad_r2)
plt.plot(X_fit, y_cubic_fit, 'm-', label='cubic(d=3), $R^2=%.2f$' % cubic_r2)
plt.legend()
```

```python
X_log = np.log(X)
y_sqrt = np.sqrt(y)

X_fit = np.arange(np.log(X.min()), np.log(X.max())+1, 1).reshape(-1, 1)
regr.fit(X_log, y_sqrt)

plt.scatter(X_log, y_sqrt, c='lightgray', label='train data')
plt.plot(X_fit, regr.predict(X_fit), 'r-')
```

```python
# random forest
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

X = df.iloc[:, :-1].values
y = df['MEDV'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)

forest = RandomForestRegressor(n_estimators=1000, criterion="mse", random_state=1, n_jobs=-1)
forest.fit(X_train, y_train)
y_train_pred = forest.predict(X_train)
y_test_pred = forest.predict(X_test)

print(mean_squared_error(y_train, y_train_pred), mean_squared_error(y_test, y_test_pred))
print(r2_score(y_train, y_train_pred), r2_score(y_test, y_test_pred))
```
