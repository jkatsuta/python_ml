## S11.1 k-means

```python
%matplotlib inline

from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=150,
                                 n_features=2,
                                 centers=3,
                                 cluster_std=0.5,
                                 shuffle=True,
                                 random_state=0)
plt.plot(X[:, 0], X[:, 1], 'o')
plt.grid()
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3,
                        #init='random',
                        init='k-means++',
                        n_init=10,
                        max_iter=300,
                        tol=1e-4,
                        random_state=0)
y_km = km.fit_predict(X)
y_km
```

```python
COLORS = list('rgb')

for c in range(3):
    plt.scatter(X[y_km==c, 0], X[y_km==c, 1],
                       marker='o', color=COLORS[c], label='cluster %d' % c)
plt.scatter(km.cluster_centers_[:, 0],
                    km.cluster_centers_[:, 1], s=200, 
                    marker='*', c='magenta', label='centroids')
plt.legend(loc="best")
```

```python
n_clusters = range(1, 11)

distortions = []
for i in n_clusters:
    km = KMeans(n_clusters=i,
                            init='k-means++',
                            n_init=10,
                            max_iter=300,
                            random_state=0)
    km.fit(X)
    distortions.append(km.inertia_)
plt.plot(n_clusters, distortions, 'o-')
```

```python
nc = 4
km = KMeans(n_clusters=nc,
                            init='k-means++',
                            n_init=10,
                            max_iter=300,
                            random_state=0)
y_km = km.fit_predict(X)

COLORS = list('rgbk')

for c in range(nc):
    plt.scatter(X[y_km==c, 0], X[y_km==c, 1],
                       marker='o', color=COLORS[c], label='cluster %d' % c)
plt.scatter(km.cluster_centers_[:, 0],
                    km.cluster_centers_[:, 1], s=200, 
                    marker='*', c='magenta', label='centroids')
plt.legend(loc="best")
```

```python
from matplotlib import cm
from sklearn.metrics import silhouette_samples

nc = 3
km = KMeans(n_clusters=nc,
                            init='k-means++',
                            n_init=10,
                            max_iter=300,
                            random_state=0)
y_km = km.fit_predict(X)

cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]

silhouette_vals = silhouette_samples(X, y_km, metric='euclidean')

y_ax_lower, y_ax_upper = 0, 0
yticks = []
for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km==c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals)
    color = cm.jet(float(i) / n_clusters)
    plt.barh(range(y_ax_lower, y_ax_upper),
                   c_silhouette_vals,
                   color=color,
                   height=1.0,
                   edgecolor='none')
    yticks.append(0.5 * (y_ax_lower + y_ax_upper))
    y_ax_lower = y_ax_upper

silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg, color='red', ls='--')
plt.yticks(yticks, cluster_labels + 1)
```

## S11.2 hierarchical clustering

```python
np.random.seed(123)
variables = list('XYZ')
labels = ['ID%d' % i for i in range(5)]
X = np.random.random_sample([len(labels), len(variables)]) * 10
df = pd.DataFrame(X, columns=variables, index=labels)
df
```

```python
from scipy.spatial.distance import pdist, squareform

d = pdist(df, metric='euclidean')
print(d)
row_dist = pd.DataFrame(squareform(d), columns=labels, index=labels)
row_dist
```

```python
from scipy.cluster.hierarchy import linkage

row_clusters = linkage(d, method='complete')
pd.DataFrame(row_clusters,
                          columns=['row1', 'row2', 'distance', 'no. of items in clust.'],
                          index=['cluster%d' % (i+1) for i in range(row_clusters.shape[0])])
```

```python
from scipy.cluster.hierarchy import dendrogram

row_dendr = dendrogram(row_clusters, labels=labels)
```

```python
fig = plt.figure(figsize=(8, 8), facecolor='white')
axd = fig.add_axes([0.09, 0.1, 0.2, 0.6])

row_dendr = dendrogram(row_clusters, orientation='left')

df_rowclust = df.ix[row_dendr['leaves'][::-1]]
axm = fig.add_axes([0.23, 0.1, 0.6, 0.6])
cax = axm.matshow(df_rowclust, interpolation='none', cmap='hot_r')

axd.set_xticks([])
axd.set_yticks([])
for i in axd.spines.values():
    i.set_visible(False)

fig.colorbar(cax)
axm.set_xticklabels([' '] + list(df_rowclust.columns))
axm.set_yticklabels([' '] + list(df_rowclust.index))
```

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=2,
                                                    affinity='euclidean',
                                                    linkage='complete')
labels = ac.fit_predict(X)
labels
```

## S11.3 DBSCAN

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200, noise=0.05, random_state=0)
plt.scatter(X[:, 0], X[:, 1])
```

```python
from sklearn.cluster import DBSCAN

fig, axes = plt.subplots(1, 3, figsize=(16, 4))

algs = [KMeans(n_clusters=2, random_state=0),
             AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='complete'),
             DBSCAN(eps=0.2, min_samples=5, metric='euclidean')]
ALGS = ['k-means', 'hierarchy', 'DBSCAN']

for i, alg in enumerate(algs):
    y_pred = alg.fit_predict(X)
    axes[i].scatter(X[y_pred==0, 0], X[y_pred==0, 1], color='b', label='cluster1')
    axes[i].scatter(X[y_pred==1, 0], X[y_pred==1, 1], color='g', label='cluster2')
    axes[i].set_title(ALGS[i])
axes[i].legend()
```
