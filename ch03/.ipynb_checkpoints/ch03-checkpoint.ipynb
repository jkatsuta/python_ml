```python
%matplotlib inline
from sklearn.datasets import load_iris
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import StandardScaler

iris = load_iris()
X = iris.data[:, [2, 3]]
y = iris.target

X_train, X_test, y_train, y_test =\
    train_test_split(X, y, test_size=0.3, random_state=0)

X_combined = np.vstack((X_train, X_test))
y_combined = np.hstack((y_train , y_test))

sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
X_combined_std = np.vstack((X_train_std, X_test_std))
```

```python
from matplotlib.colors import ListedColormap

def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):
    markers = ['s', '^', 'o', 'v', 'x']
    colors = ['red', 'blue', 'lightgreen', 'gray', 'cyan']
    cmap = ListedColormap(colors[: len(np.unique(y))])
    
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                                              np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    
    plt.figure(figsize=(12, 8))
    plt.contourf(xx1, xx2, Z, alhpa=0.2, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    for i, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0],
                            y=X[y == cl, 1],
                            alpha=0.8, c=cmap(i),
                            marker=markers[i], label=cl)
    if test_idx:
        X_test, y_test = X[test_idx, :], y[test_idx]
        plt.scatter(X_test[:, 0], X_test[:, 1], 
                           c='', alpha=1.0, marker='o', s=55, 
                           linewidths=1, label='test set')
    plt.legend(loc='best')
```

## 3.3.3 Logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=1000.0)
lr.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, y_combined, classifier=lr, test_idx=range(105, 150))
```

## 3.4 Support vector machine

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear', C=1.)
svm.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, y_combined, classifier=svm, test_idx=range(105, 150))
```

## 3.5 kernel SVM

```python
np.random.seed(0)
X_xor = np.random.randn(200, 2)
y_xor = np.logical_xor(X_xor[:, 0] > 0, X_xor[:, 1] > 0)
y_xor = np.where(y_xor, 1, -1)

plt.figure(figsize=(8, 6))
plt.scatter(X_xor[y_xor==1, 0], X_xor[y_xor==1, 1], marker='x', label='1', c='b')
plt.scatter(X_xor[y_xor==-1, 0], X_xor[y_xor==-1, 1], marker='s', label='-1', c='r')
```

```python
svm = SVC(kernel='linear')
svm.fit(X_xor, y_xor)
plot_decision_regions(X_xor, y_xor, classifier=svm)

svm = SVC(kernel='rbf', gamma=0.1, C=10.)
svm.fit(X_xor, y_xor)
plot_decision_regions(X_xor, y_xor, classifier=svm)
```

```python
svm = SVC(kernel='rbf', gamma=0.2, C=1.)
svm.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, y_combined, classifier=svm)
```

```python
svm = SVC(kernel='rbf', gamma=100., C=1.)
svm.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, y_combined, classifier=svm)
```

## 3.6 decision tree / random forest

```python
def gini(p):
    return p * (1 - p)  + (1- p) * p

def entropy(p):
    return -(p * np.log2(p) + (1 - p) * np.log2(1 - p))

def error(p):
    return 1 - np.maximum(p, 1-p)

%matplotlib inline
eps = 1e-6
x = np.linspace(eps, 1-eps, 100)

for f in [gini, entropy, error]:
    plt.plot(x, f(x), '-')
plt.plot(x, 0.5 * entropy(x), 'g--')
plt.legend()
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion='entropy',
                                                   max_depth=3,
                                                   random_state=0)

tree.fit(X_train, y_train)
plot_decision_regions(X_combined, y_combined, classifier=tree, test_idx=range(105, 150))
```

## 3.6.3

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(criterion='entropy', n_estimators=10, random_state=1, n_jobs=2)
forest.fit(X_train, y_train)
plot_decision_regions(X_combined, y_combined, classifier=forest, test_idx=range(105, 150))
```

## 3.7 knn

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5, p=2, metric='minkowski')
knn.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, y_combined, classifier=knn, test_idx=range(105, 150))
```
